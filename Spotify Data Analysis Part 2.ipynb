```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
s_tracks = pd.read_csv('tracks.csv')
```

```python
s_tracks.head()
```

```python
s_tracks.set_index("release_date", inplace=True)
s_tracks.index=pd.to_datetime(s_tracks.index)
s_tracks.head()
```

```python
s_tracks['dates']=s_tracks.index.get_level_values('release_date')
s_tracks.dates=pd.to_datetime(s_tracks.dates)
years=s_tracks.dates.dt.year
```

```python
total_pop = s_tracks.popularity
fig_pop = (18,7)
fig, ax = plt.subplots(figsize = fig_pop)
fig = sns.barplot(x = years ,
                 y = total_pop,
                 ax = ax,
                 errwidth = False).set(title="Years  vs popularity")
plt.xticks(rotation=90)
```
