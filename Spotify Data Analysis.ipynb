#                                              #Corelation heatmap Visual

```python
corr_s=s_tracks.drop(["key","mode","explicit"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_s,
                     annot=True,
                     fmt=".1g", 
                     vmin=-1, 
                     vmax=1,
                     center=0,
                     cmap="inferno",
                     linewidths=1,
                     linecolor="Black")
heatmap.set_title("Correlation Heatmap B/W Variable")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=60)
```

```python
sample_s = s_tracks.sample(int(0.004*len(s_tracks)))
print(len(sample_s))
```

# Regraresion Plot

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_s,
           y = "loudness",
           x = "energy",
           color = "c").set(title = "Loudness Vs Energy Corelation plot")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_s,
           y = "popularity",
           x = "energy",
           color = "c").set(title = "Popularity Vs Energy Corelation plot")
```

```python
s_tracks['dates']=s_tracks.index.get_level_values('release_date')
s_tracks.dates=pd.to_datetime(s_tracks.dates)
years=s_tracks.dates.dt.year
```

# Histogrm Plot

```python
sns.displot(years,discrete=True,
            aspect=2,
            height=5,
            kind="hist").set(title="Num of songs per year")
```

```python
total_dur = s_tracks.duration
fig_dim = (18,7)
fig, ax = plt.subplots(figsize = fig_dim)
fig = sns.barplot(x = years ,
                 y = total_dur,
                 ax = ax,
                 errwidth = False).set(title="Years  vs Duration")
plt.xticks(rotation=90)
```

```python
total_pop = s_tracks.popularity
fig_pop = (18,7)
fig, ax = plt.subplots(figsize = fig_pop)
fig = sns.barplot(x = years ,
                 y = total_pop,
                 ax = ax,
                 errwidth = False).set(title="Years  vs popularity")
plt.xticks(rotation=90)
```
